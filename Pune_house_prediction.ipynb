```python
import pandas as pd
import numpy as np
```

```python
df_cleaned = pd.read_csv("/Pune house data.csv")
```

CLEANED TOTAL_SQFT COLUMN

```python
df_cleaned['total_sqft'] = df_cleaned['total_sqft'].str.replace(r'[^\d-]', '', regex=True)
df_cleaned['total_sqft'] = df_cleaned['total_sqft'].apply(lambda x: int(sum(map(int, x.split('-')))/2) if '-' in x else int(x))
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

CLEANED BHK COLUMN

```python
# Extract the number of BHK from the 'size' column and rename the column
df_cleaned['bhk_final'] = df_cleaned['size'].str.extract('(\d+)').fillna(0).astype(float).astype(pd.Int64Dtype())
```

```python
df_cleaned = df_cleaned.drop(columns=['size'])
```
